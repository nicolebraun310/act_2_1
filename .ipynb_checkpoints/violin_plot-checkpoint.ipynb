```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
# filter down to only cars with even number of cylinders
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.violinplot(cars.displacement);
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars
              );
```

```python
# these violin plots are symmetric, however that does not convey any additional information
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])]
              )
plt.legend(loc=2);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               # can only use split when the hue has only 2 properties
               split=True
              )
plt.legend(loc=2);
```

```python
#can also change how the inner boxes look
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               inner = 'quartile'
              )
plt.legend(loc=2);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               # each violin has the exact same amount of area before scaling
               scale='count'
              )
plt.legend(loc=2);
```

```python
(
    cars[cars.origin.isin(['japan', 'europe'])]
    .groupby('cylinders')
    .origin
    .value_counts()
)
```

```python
# styling options
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
               order=['japan', 'europe', 'usa']
              );
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
               linewidth=3
              );
```

```python
# KDE options
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement,
               # changing band width
               bw=0.2
              );
```
